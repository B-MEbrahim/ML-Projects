```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('online_shoppers_intention.csv')
df.head()
```

```python
df.isna().sum()
```

```python
X = df.iloc[:, [5, 6]].values
X.shape
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42, tol=0.001)
    km.fit(X)
    labels = km.labels_
    wcss.append(km.inertia_)


plt.rcParams['figure.figsize'] = (13, 7)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_means = km.fit_predict(X)

plt.scatter(X[y_means == 0, 0], X[y_means == 0, 1], s=50, c='yellow', label='Uninterested Customer')
plt.scatter(X[y_means == 1, 0], X[y_means == 1, 1], s=50, c='red', label='Target Customer')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=50, c='blue', label='Centeroid')

plt.title('ProductRelated Duration vs Bounce Rate', fontsize = 20)
plt.grid()
plt.xlabel('ProductRelated Duration')
plt.ylabel('Bounce Rates')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
labels_true = le.fit_transform(df['Revenue'])

labels_pred = y_means

from sklearn import metrics

score = metrics.adjusted_rand_score(labels_true, labels_pred)
print(score)
```
